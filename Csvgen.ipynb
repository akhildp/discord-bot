```python
import pandas as pd
import requests
```

```python
ENDPOINT = 'https://api.coingecko.com/api/v3/coins/markets?vs_currency=usd&ids='
```

```python
df = pd.read_csv('coins_endpoints.csv')
```

```python
df = df[['Name']]
```

```python
df['Symbol'] = [requests.get(ENDPOINT+coin).json()[0]['symbol'] for coin in df['Name']]
```

```python
df['ATH'] = [requests.get(ENDPOINT+coin).json()[0]['ath'] for coin in df['Name']]
```

```python
df['ATH_PC'] = [requests.get(ENDPOINT+coin).json()[0]['ath_change_percentage'] for coin in df['Name']]
```

```python
df.to_csv('coins_endpoints.csv', index='False')
```
